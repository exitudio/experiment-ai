```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import math
```

### Geometric Series

```python
gamma = 0.9
n = 3

def geometric_series_infinite(gamma):
    return 1/(1-gamma)
def geometric_series_partial(n, gamma):
    total = 0
    for i in range(n):
        total += 1*gamma**i
    return total

# infinite
print((1-gamma) * geometric_series_infinite(gamma))

# partial
print(
    (1-gamma) * geometric_series_partial(n, gamma) + 
    gamma**n
)
```

### Cross Entropy

```python
plt.style.use('seaborn-whitegrid')
fig = plt.figure()
ax = plt.axes()

p = np.linspace(.0001, .9999, 100)
one = -p * np.log(p)
zero = -(1-p) * np.log(1-p)
ax.plot(p, one);
ax.plot(p, zero);
ax.plot(p, one + zero);
```

### log_prob
just wanna prove log_prob is just the log of normal pdf
[here](https://pytorch.org/docs/stable/_modules/torch/distributions/normal.html#Normal.log_prob)

```python
mu = 0+1e-10
sd = 1

def normpdf(x, mu, sd):
    var = sd**2
    denom = (2*math.pi*var)**.5
    num = np.exp(-(x-mu)**2/(2*var))
    return num/denom
def log_prob(x, mu, sd):
    var = sd**2
    log_scale = math.log(sd)
    return -((x - mu) ** 2) / (2 * var) - log_scale - math.log(math.sqrt(2 * math.pi))

x = np.linspace(-5, 5, 100)
plt.plot(x, log_prob(x, mu, sd))
plt.plot(x, np.log(normpdf(x, mu, sd)))
```
