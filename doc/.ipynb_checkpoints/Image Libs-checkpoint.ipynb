## Image Transformation

### 1. AffineTransform

````python
img = cv2.imread('display_data/test.png')
rows,cols,ch = img.shape
pts1 = np.float32([[0,0],[cols,0],[0,rows]])
pts2 = np.float32([[0,0],[cols+30,20],[20,rows+30]])

M = cv2.getAffineTransform(pts1,pts2)

dst = cv2.warpAffine(img,M,(cols+100,rows+100))

imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
dstRGB = cv2.cvtColor(dst, cv2.COLOR_BGR2RGB)
plt.subplot(121),plt.imshow(imgRGB),plt.title('Input')
plt.subplot(122),plt.imshow(dstRGB),plt.title('Output')
plt.show()
````

### 2. Perspective Transformation

````python
img = cv2.imread('display_data/test.png', cv2.IMREAD_UNCHANGED)
cols, rows, ch = img.shape

pts1 = np.float32([[0,0],[cols,0],[0,rows],[cols, rows]])
pts2 = np.float32([[0,0],[cols-10,20],[20,rows-10],[cols,rows]])

M = cv2.getPerspectiveTransform(pts1,pts2)

# cv2.warpPerspective( img, getPerspectiveTransform, (cols, rows))
dst = cv2.warpPerspective(img,M,(rows+100, cols+100))

imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
dstRGB = cv2.cvtColor(dst, cv2.COLOR_BGR2RGB)
plt.subplot(121),plt.imshow(imgRGB),plt.title('Input')
plt.subplot(122),plt.imshow(dstRGB),plt.title('Output')
plt.show()
dst.shape
````

### convert to RGB with alpha channel
- plt.imshow - RGB, cv2 - BGR
- because cv2.cvtColor will delete alpha channel
- be careful, type of plt.imshow has to be int

````python
temp = np.zeros(dst.shape).astype(np.uint8)
temp[:, :, 0] = dst[:, :, 2]
temp[:, :, 1] = dst[:, :, 1]
temp[:, :, 2] = dst[:, :, 0]
temp[:, :, 3] = dst[:, :, 3]

plt.imshow(temp)
````

## Copy Image to Image

````python
x_offset = 100
y_offset = 100
s_img = temp
l_img = cv2.imread('display_data/background.png', cv2.IMREAD_UNCHANGED)

y1, y2 = y_offset, y_offset + s_img.shape[0]
x1, x2 = x_offset, x_offset + s_img.shape[1]

alpha_s = s_img[:, :, 3] / 255.0
alpha_l = 1.0 - alpha_s

for c in range(0, 3):
    l_img[y1:y2, x1:x2, c] = (alpha_s * s_img[:, :, c] +
                              alpha_l * l_img[y1:y2, x1:x2, c])
plt.imshow(l_img)
````

### HSV

````python
img = np.array([[[255, 0, 0], [255, 0, 0] ],
                [[0, 255, 0], [0, 0, 255]]])
img = img.astype(np.uint8)
img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

index = 141
def plot(img, h):
    global index
    img[0][0][0] = h
    print('hsv:', img[0][0])
    img = cv2.cvtColor(img, cv2.COLOR_HSV2BGR)
    plt.subplot(index),plt.imshow(img),plt.title('Input')
    index += 1
    
plot(img.copy(), 0) # blue
plot(img.copy(), 60) # green
plot(img.copy(), 120) # red
plot(img.copy(), 180) # blue
````

### Convert BGR to HSV with Alpha Channel
[link](https://stackoverflow.com/questions/32290096/python-opencv-add-alpha-channel-to-rgb-image)

#### Solution 1 : Merge 
```python
image = cv2.merge((r,g,b,a))
```
#### Solution 2 : Add alpha channel
```python
rgba = cv2.cvtColor(rgb_data, cv2.COLOR_RGB2RGBA)
rgba[:, :, 3] = alpha_data
```

#### Step by step
1. save alpha channel 
2. convert to HSV
3. do what ever you wanna do
4. convert back to BGR 
5. merge(B, G, R, A)
```python
temp_alpha = image[:, :, 3]
image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
rand = np.random.randint(0, 180)
image[:, :, 0] += rand
image = cv2.cvtColor(image, cv2.COLOR_HSV2BGR)
# convert back with alpha channel using merge or add alpha channel
image = cv2.merge((
    image[:, :, 0], 
    image[:, :, 1], 
    image[:, :, 2], 
    temp_alpha))
```

### Rotate Point

````python
def rotate_image(image, pivot, degrees):
    rows, cols, _ = image.shape
    M = cv2.getRotationMatrix2D(pivot,degrees,1)
    image = cv2.warpAffine(image,M,(cols, rows))
    return image, M
    
def get_rotated_point(rotation_matrix, points):
    ones = np.ones(shape=(len(points), 1))
    points_ones = np.hstack([points, ones])
    rotated_point = rotation_matrix.dot(points_ones.T).T
    return rotated_point

pivot = (cols/2,rows/2)
dot = (30, 30)
image = cv2.imread('display_data/test.png')
cv2.circle(image, dot, 2, (255, 0, 0), 1)
image, M = rotate_image(image, pivot, 45)
plt.imshow(image)
get_rotated_point(M, np.array([dot]))
````

````python
import math
def get_rotate_matrix(border_points, radian_angle, pivot, offset_x=0, offset_y=0):
    scale = 1
    x, y = pivot
    a = scale*math.cos(radian_angle)
    b = scale*math.sin(radian_angle)
    M = np.float32([[a, b, (1-a)*x - b*y + offset_x],
                    [-b, a, b*x + (1-a)*y + offset_y]])
    rotated_points = get_rotated_point(M, border_points)
    rotated_points = np.array(rotated_points)
    min_point = np.min(rotated_points, 0)
    max_point = np.max(rotated_points, 0)
    return M, min_point, max_point

    
def rotate_bound(image, angle):
    """Mimic cv2.getRotationMatrix2D"""
    radian_angle = angle/180*math.pi
    rows, cols, _ = image.shape
    border_points = ((0,0), (cols, 0), (0, rows), (cols, rows))
    pivot = (rows/2, cols/2)
    
    M, min_point, max_point = get_rotate_matrix(border_points, radian_angle, pivot)
    
    if min_point[0] != 0 or min_point[1] != 0:
        # move image if necessary
        offset_x = -min_point[0]
        offset_y = -min_point[1]
        M, min_point, max_point = get_rotate_matrix(border_points, radian_angle, pivot, offset_x, offset_y)
    
    image = cv2.warpAffine(image,M,(int(max_point[0]), int(max_point[1])))
    return image

image = cv2.imread('display_data/test.png')
image = rotate_bound(image, 45)
plt.imshow(image)
````

````python
def rotate_bound(image, angle):
    rows, cols, _ = image.shape
    rotation_matrix = cv2.getRotationMatrix2D((cols/2, rows/2), angle, 1)
    rotated_points = get_rotated_point(
        rotation_matrix,
        ((0,0), (cols, 0), (0, rows), (cols, rows))
    )
    print(rotated_points)
image = cv2.imread('display_data/test.png')
rotate_bound(image, 45)
````
