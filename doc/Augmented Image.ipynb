```python
from imgaug import augmenters as iaa
```

```python
def display_effect(effect):
    image = cv2.imread('display_data/bg.png', cv2.IMREAD_UNCHANGED)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.subplot(121)
    plt.imshow(image)
    plt.title('original')
    
    seq = iaa.Sequential([effect])
    image = seq.augment_images([image])[0]
    plt.subplot(122)
    plt.imshow(image)
    plt.title('effect')
    
```

```python
display_effect( iaa.Sharpen(alpha=(0, 1.0), lightness=(0.75, 1.5)) )
```

```python
display_effect( iaa.AdditiveGaussianNoise(
                    loc=0, scale=(0.0, 0.05*255), per_channel=0.5
                ))
```

### Blur

```python
display_effect( iaa.GaussianBlur(sigma=(0, 3.0)) )
```

```python
display_effect( iaa.AverageBlur(k=(2, 10)) )
```

```python
display_effect( iaa.MedianBlur(k=(3, 11)) )
```

### Dropout

```python
display_effect( iaa.Dropout((0.01, 0.1), per_channel=0.5) )
```

```python
display_effect( iaa.CoarseDropout(
                    (0.03, 0.15), size_percent=(0.02, 0.05),
                    per_channel=0.2
                ) )
```

### Change color

```python
display_effect( iaa.Add((-40, 40), per_channel=0.5) )
```

```python
display_effect( iaa.Multiply((0.7, 1.3), per_channel=0.5) )
```

```python
display_effect( iaa.ContrastNormalization((0.5, 2.0), per_channel=0.5) )
```

```python
display_effect( iaa.AddToHueAndSaturation((-40, 40)) )
```

### Distort

```python
display_effect( iaa.ElasticTransformation(alpha=(0.5, 3.5), sigma=0.25) )
```
