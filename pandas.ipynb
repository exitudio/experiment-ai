### group

```python
groups = diabetes.groupby('Group').groups
for g in groups:
    print(g)
```

### normalize

```python
col = list(diabetes.drop('Group',axis=1).columns)
diabetes[col] = diabetes[col].apply(lambda x: (x - x.min()) / (x.max()-x.min()))
diabetes[col].head()
```
