```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

```python
from exit import NeuralNetwork
from exit.layers import Dense
from exit.activations import Sigmoid, Relu
from exit.losses import Cross_entropy, L2
```

### Regression

```python
nn = NeuralNetwork()
input = np.linspace(0,20,21).reshape(-1,1)
output = (input*2+10)
nn.sequence(
    input.shape[1],
    Dense(4, Relu),
    Dense(1),
)
for i in range(100):
    predict = nn.train(input, output, L2, learning_rate=0.00003)
    if i%99 is 0:
        print(f'loss {i} :', predict['loss'])
```

```python
plt.plot(input, output)
plt.plot(input, nn.predict(input))
```

### Classifiaction

```python
nn = NeuralNetwork()
input = np.linspace(0,20,100).reshape(-1,1)
output = (input*2+10 > 30).astype(int)
nn.sequence(
    input.shape[1],
    Dense(1, Sigmoid),
)
for i in range(1000):
    predict = nn.train(input, output, Cross_entropy, learning_rate=0.01)
    if i%99 is 0:
        print(f'loss: {i} = ', predict['loss'])
```

```python
plt.plot(input, output)
plt.plot(input, nn.predict(input))
```
