### 0. Helper

```python
def plot_image(image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image)
```

```python
def overlay_image(top_image, background_image, x_offset, y_offset):
    y1 = y_offset
    y2 = y_offset + top_image.shape[0]
    x1 = x_offset
    x2 = x_offset + top_image.shape[1]
    
    # if top image go futher that the end of background image
    if y2 > background_image.shape[0]:
        cut_image_height = top_image.shape[0] - (y2 - background_image.shape[0])
        y2 = background_image.shape[0]
        top_image = top_image[:cut_image_height, :, :]
    if x2 > background_image.shape[1]:
        cut_image_width = top_image.shape[1] - (x2 - background_image.shape[1])
        x2 = background_image.shape[1]
        top_image = top_image[:, :cut_image_width, :]

    
    alpha_top = top_image[:, :, 3] / 255.0
    alpha_background = 1.0 - alpha_top
    image = background_image[:,:,:]

    for c in range(0, 3):
        a1 = alpha_top * top_image[:, :, c]
        a2 = alpha_background * background_image[y1:y2, x1:x2, c]
        image[y1:y2, x1:x2, c] = (a1 + a2)
    return image
```

### 1. Gen background image

```python
def gen_background(image_path, image_size):
    image = cv2.imread(image_path, cv2.IMREAD_UNCHANGED)
    cols, rows, _ = image.shape
    x = np.random.randint(0, cols - image_size)
    y = np.random.randint(0, rows - image_size)
    image = image[x:x+image_size, y:y+image_size]
    return image
```

```python
image = gen_background('source-data/background/bg1.jpg', 224)
plot_image(image)
```

### 2. Gen perspective image

```python
image_manipulation = Image_manipulation.Image_manipulation()
def gen_distort_image(
        image,
        output_corners,
        width, 
        height):

    # declare input corners
    input_width, input_height, ch = image.shape
    input_corners = np.float32([[0, 0],
                               [input_width, 0],
                               [0, input_height],
                               [input_width, input_height]])

    # transform
    perspectiveTransform = cv2.getPerspectiveTransform(input_corners, output_corners)
    image = cv2.warpPerspective(image, perspectiveTransform, (width, height))
    
    angle = np.random.randint(-180, 181)
    image, rotated_output_corners = image_manipulation.rotate_bound(image, angle, output_corners)
    
    # draw bounding box [for debugging]
#     lines = [np.int32([rotated_output_corners[0], 
#                        rotated_output_corners[1], 
#                        rotated_output_corners[3], 
#                        rotated_output_corners[2]])]
#     cv2.polylines(image, lines, True, color=(0, 0, 255, 255), thickness=1)
    
    return image, rotated_output_corners
```

```python
def random_corners(width, height, small_bound_scale):
    # small bound
    half_small_bound_scale = small_bound_scale*.5
    SMALL_RECT_LEFT = width* (.5-half_small_bound_scale)
    SMALL_RECT_TOP = height* (.5-half_small_bound_scale)
    SMALL_RECT_RIGHT = width* (.5+half_small_bound_scale)
    SMALL_RECT_BOTTOM = height* (.5+half_small_bound_scale)
    
    # random all points in bound
    left = np.random.randint(0, SMALL_RECT_LEFT, size=2)
    right = np.random.randint(SMALL_RECT_RIGHT, width, size=2)
    top = np.random.randint(0, SMALL_RECT_TOP, size=2)
    bottom = np.random.randint(SMALL_RECT_BOTTOM, height, size=2)
    
    top_left = [left[0], top[0]]
    right_top = [right[0], top[1]]
    left_bottom = [left[1], bottom[0]]
    right_bottom = [right[1], bottom[1]]
    return np.float32([top_left, right_top, left_bottom, right_bottom])
```

```python
max_width = max_height = 224
width = np.random.randint( .3 * max_width, max_width)
height = np.random.randint( .3 * max_height, max_height)
output_corners = random_corners(width, height, .5)
object_image = cv2.imread('source-data/object/dt0.png', cv2.IMREAD_UNCHANGED)
image, corners = gen_distort_image(object_image, output_corners, width, height)

plot_image(image)
```

### 3. Gen object with background

```python
IMAGE_SIZE = 224
ICON_MIN_SIZE = 60
ICON_MAX_SIZE = 200
SMALL_BOUND_SCALE = .5

def gen_object_with_background(object_image_path, background_image_path):
    width = np.random.randint( ICON_MIN_SIZE, ICON_MAX_SIZE)
    height = np.random.randint( ICON_MIN_SIZE, ICON_MAX_SIZE)
    output_corners = random_corners(width, height, small_bound_scale=SMALL_BOUND_SCALE)
    
    # gen images
    object_image = cv2.imread(object_image_path, cv2.IMREAD_UNCHANGED)
    object_image, corners = gen_distort_image(object_image, output_corners, width, height)
    bg_image = gen_background(background_image_path, IMAGE_SIZE)
    
    obj_rows, obj_cols, _ = object_image.shape
    bg_rows, bg_cols, _ = bg_image.shape

    # add offseta
    offset_x = np.random.randint(0, bg_cols - obj_cols) if bg_cols - obj_cols > 0 else 0
    offset_y = np.random.randint(0, bg_rows - obj_rows) if bg_rows - obj_rows > 0 else 0
    corners[:, 0] += offset_x
    corners[:, 1] += offset_y

    image = overlay_image(object_image, bg_image, offset_x, offset_y)
    return image, corners
```

```python
image, corners = gen_object_with_background('source-data/object/dt0.png', 'source-data/background/bg1.jpg')
plot_image(image)
```

### 4. Augmented data

```python
from imgaug import augmenters as iaa
seq = iaa.Sequential([
    iaa.SomeOf(1, [
        iaa.GaussianBlur(sigma=(0, 3.0)),
        iaa.AverageBlur(k=(2, 10)),
        iaa.MedianBlur(k=(3, 11)),
        iaa.Dropout((0.01, 0.1), per_channel=0.5),
        iaa.CoarseDropout(
            (0.03, 0.15), size_percent=(0.02, 0.05),
            per_channel=0.2
        ),
        iaa.ElasticTransformation(alpha=(0.5, 3.5), sigma=0.25)
    ]),
    iaa.Add((-10, 10), per_channel=0.5),
    iaa.Multiply((0.7, 1.3), per_channel=0.5),
    iaa.AddToHueAndSaturation((-5, 5), per_channel=1)
])
augmented_image = seq.augment_images
images_aug = augmented_image([image])
plot_image(images_aug[0])
```

### 5. Gen all images

```python
BACKGROUND_FOLDER = 'source-data/background'
ICON_PATH = 'source-data/object/dt0.png'

def gen_images(num_image, path, is_augmented=True):
    if os.path.isdir(path):
        shutil.rmtree(path)
    os.mkdir(path)
    
    background_image_paths = get_files_in_folder(BACKGROUND_FOLDER)
    corners = []
    for i in range(num_image):
        background_image_index = np.random.randint(0, len(background_image_paths))
        background_path = background_image_paths[background_image_index]
        image, one_image_corners = gen_object_with_background(ICON_PATH, f'{BACKGROUND_FOLDER}/{background_path}')
        
        # debug bound
#         cv2.polylines(image, 
#                       [np.int32([one_image_corners[0],
#                                   one_image_corners[1],
#                                   one_image_corners[3],
#                                   one_image_corners[2]])], 
#                       True, 
#                       color=(0, 0, 255, 255), 
#                       thickness=1)
        
        # concatenate all points to one array
        one_image_corners = np.concatenate(one_image_corners) 
        corners.append(one_image_corners)
        
        # augmented image
        image = augmented_image([image])[0] if is_augmented else image
        
        # save image
        cv2.imwrite(f'{path}/object_image{i}.png',image)
    
    # save corners
    file = open(f'{path}/corners.csv', 'w')
    with file:
        writer = csv.writer(file)
        writer.writerows(corners)
```

```python
TRAIN_FOLDER = './data/train'
VAL_FOLDER = './data/val'
TEST_FOLDER = './data/test'

gen_images(8224, TRAIN_FOLDER)
gen_images(512, TEST_FOLDER)
gen_images(512, VAL_FOLDER)
```
