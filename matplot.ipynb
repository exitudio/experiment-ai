```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

data = np.random.normal(0,1,10000)
y, binEdges = np.histogram(data, bins=100)
bincenters = 0.5*(binEdges[1:]+binEdges[:-1])
plt.plot(bincenters,y,'-')
```
